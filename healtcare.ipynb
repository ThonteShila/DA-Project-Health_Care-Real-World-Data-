```python
def consult(result):
    if result=="Normal":
        return "Not necessary" 
    elif result=="Inconclusive":
        return  "Meet Dr every 3 month"
    elif result=="Abnormal":
        return "Need Attention"
    else:
        "Must visit Dr"

df["Action Needed"]=df["Test Results"].apply(consult)
```

```python
df
```

```python

df["Admission Type"].value_counts()
```

```python
def cases(result):
    if result=="Elective":
        return "not immediately"
    elif result=="Urgent":
        return "require attention"
    else: 
        return " Immediately"
df["Case_type"]=df["Admission Type"].apply(cases)
df
```
